### Etapa 4: Balanceamento dos Dados
- Para efetuar o balanceamento dos dados, utilizarei a biblioteca **imblearn**, que oferece uma abordagem eficiente para equilibrar as distribuições de dados. Antes de proceder com o balanceamento, dividirei os dados em conjuntos de características (X) e alvo (Y), e também realizarei a separação entre dados de treino e teste. Essa estratégia ajudará a tornar o processo de balanceamento mais eficaz e alinhado com as necessidades do modelo.

```python
#Separando meus dados Entre X e y e entre dados de treino e teste

# Definição da Variável para Identificação de Y
x = df.drop('Class',axis=1)
y = df['Class']

# Divisão dos Dados em Conjuntos de Treino e Teste
x_train, x_test, y_train, y_test = train_test_split(x,y, random_state=10)
```

```python
# Aplicação da Técnica de Under-Sampling
rus = RandomUnderSampler()

#Treinando o modelo que irá fazer esse balanceamento
x_rus, y_rus = rus.fit_resample(x_train,y_train)

#Exibindo em forma grafica
sns.countplot(y_rus)
```

---

- E com isso, com os gráficos devidamente equilibrados, estamos prontos para avançar na implementação do nosso modelo. Agora, ele estará apto a identificar padrões de fraude, permitindo a classificação precisa de atividades como fraudulentas ou não.

### Etapa 5: "Implementação e Avaliação do Melhor Modelo de Aprendizado de Máquina"

```python
#Criando uma Função para Avaliar e Testar os Modelos.

def Avaliarmodelo(nome_modelo, y_test, previsao):
    sk = skplt.metrics.plot_confusion_matrix(y_test, previsao, normalize=True)
    cm = classification_report(y_test, previsao, digits=4)
    ac = accuracy_score(y_test,previsao)
    roc = roc_auc_score(y_test, previsao)
    return print(f'O Modelo: {nome_modelo}\nRelatório de Classificação:\n{cm}\nAcurácia: {ac:.4f}\nAUC: {roc:.4f}\n')
```

```python
#Instanciando os modelos em variaveis

#modelo_lc = LogisticRegression()
#modelo_dc = DecisionTreeClassifier()
modelo_rf = RandomForestClassifier()
#modelo_ex = ExtraTreeClassifier()
#modelo_nb = GaussianNB()
#modelo_kn = KNeighborsClassifier()
#modelo_svc = SVC()

# Construindo um Dicionário para Facilitar o Acesso aos Modelos"
modelos = {
    #'Logistic Regression':modelo_lc,
    #'Decision Tree Classifier':modelo_dc,
    'Random Forest Classifier':modelo_rf,
    #'Extra Tree Classifier':modelo_ex,
    #'GaussianNB':modelo_nb,
    #'KNeighbors Classifier':modelo_kn,
    #'SVC':modelo_svc
}
```

```python
#Percorrendo e Treinando Modelos na Lista de Dicionários

for nome_modelo, Modelo in modelos.items():
    #Treinando os modelos
    Modelo.fit(x_rus, y_rus)

    #Fazendo previsão com os modelos
    previsao = Modelo.predict(x_test)

    #Avaliando os modelos
    Avaliarmodelo(nome_modelo, y_test, previsao)
```

### Etapa 6: Melhorias no melhor modelo

```python
#Consolidando Todo o Código Nesta Célula para Testes Rápidos de Ajustes no Modelo
x = df.drop('Class',axis=1)
#Variável Designada para Identificar o 'y'
y = df['Class']

#Separando em Treino e Teste
x_train, x_test, y_train, y_test = train_test_split(x,y, random_state=10)

#Usando a tecnica under-sampling
rus = RandomUnderSampler()

#Treinando o modelo que irá fazer o balanceamento
x_rus, y_rus = rus.fit_resample(x_train,y_train)

#Modelo campeão, vamos colocar alguns parametros genericos para ver melhora ou piora.
modelo_rf = RandomForestClassifier(max_depth=7,n_estimators=200,max_features=14)

#Treinando o modelo campeão novamente
modelo_rf.fit(x_rus,y_rus)

#Fazendo a previsão com ele
previsao = modelo_rf.predict(x_test)

#Avaliando
Avaliarmodelo('Random Forest Classifier',y_test,previsao)
```
